```python
# Function 1
# Function 2
# Plot on 2D axis
# Plot on 3d
import numpy as np
import pylab as plt
from mpl_toolkits import mplot3d

def sin_1(x):
  return np.sin(x)

def sin_2(x):
  return np.sin(1.5 * x + 2)
  # np.sin(x +  3.14 /2) for nice circle spiral

num_timesteps = 3000
dt = 0.01
times = np.arange(0, num_timesteps * dt , dt)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 3))
ax.plot(times, sin_1(times))
_ = ax.plot(times, sin_2(times))
```

```python
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
fig = plt.figure(figsize=(21,7))
ax = fig.add_subplot(1, 2, 1, projection='3d')
p = ax.scatter(sin_1(times), sin_2(times), times, c=times, cmap="viridis")
plt.colorbar(p)
ax.set_xlim3d(-2,2)
ax.set_ylim3d(-2,2)
ax.set_zlim3d(0,50)


ax.set_xlabel('sin_1', fontsize=15)
ax.set_ylabel('sin_2', fontsize=15)
_ = ax.set_zlabel('t', fontsize=15)
```
